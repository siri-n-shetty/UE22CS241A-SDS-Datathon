#Question-7

Calculate the correlation (Using heat map) between price and all other
numeric columns and list them in descending order and identify the
highest and lowest correlation?

```python
import seaborn as sns

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_df = df[numeric_columns]

correlation_matrix = numeric_df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

correlation_values = correlation_matrix['price'].sort_values(ascending=False)

highest_correlation = correlation_values.iloc[1]  # Exclude the self-correlation with 'price'
lowest_correlation = correlation_values.iloc[-1]

print("Highest correlation:", highest_correlation)
print("Lowest correlation:", lowest_correlation)
```

#Question-6

Check for any outliers in all numeric columns and then analyze carefully,
how they should be addressed.

```python
z_scores = (df.select_dtypes(include=np.number) - df.select_dtypes(include=np.number).mean()) / df.select_dtypes(include=np.number).std()

# Find the absolute z-scores greater than 3 (considered as outliers)
outliers = np.abs(z_scores) > 3

columns_with_outliers = outliers.any()
print("Columns with outliers:")
print(columns_with_outliers)

# Analyze the outliers and decide how to address them
# Some possible approaches to address outliers include:
# 1. Removing the outliers
# 2. Replacing the outliers with a suitable value (e.g., mean, median, or a custom value)
# 3. Transforming the data (e.g., log transformation) to reduce the impact of outliers
# 4. Using robust statistical methods that are less sensitive to outliers
# The approach to address outliers depends on the specific context and the goals of the analysis.
```

#Question-4

Identify and describe any data quality issues or inconsistencies within the
data set. What steps would you take to clean and pre-processes the data to
ensure its accuracy for further analysis.

```python
import pandas as pd

data = pd.read_csv('dataset_2_Diamonds Prices.csv')

# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:")
print(missing_values)

# Handling missing values - replacing NaNs with mean for numerical columns
for col in data.select_dtypes(include=['float64', 'int64']).columns:
    try:
        data[col].fillna(data[col].mean(), inplace=True)
    except Exception as e:
        print(f"Error handling missing values in column {col}: {e}")

filter_cols = ['depth', 'table', 'price', 'x', 'y', 'z']
for col in filter_cols:
    try:
        q1 = data[col].quantile(0.25)
        q3 = data[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    except Exception as e:
        print(f"Error in column {col}: {e}")

# Encoding categorical variables - using get_dummies for one-hot encoding
data = pd.get_dummies(data, columns=['cut', 'color', 'clarity'])

# Data normalization - using Min-Max scaling for numerical columns
for col in filter_cols:
    try:
        data[col] = (data[col] - data[col].min()) / (data[col].max() - data[col].min())
    except Exception as e:
        print(f"Error in column {col}: {e}")

print("\nCleaned Dataset:")
print(data.head())

data.to_csv('cleaned_diamonds_data.csv', index=False)

# In y110, x194, z249, x316, depth390, y454, x539, z586 etc etc there are mising values, 
# these indicate data quality inconsistencies. In order to ensure that there are no such inconsistencies, 
# proper data cleaning should be ensured in order to prevent the existence of any outliers, 
# missing data can be handled using a missing data heat map, missing data percentage list or a missing data histogram. 
# Furthermore, if we are sure that the missing data is not informative, we can drop the entire observation, 
# this process is called observation dropping. 
# In the given program, we have handled it by replacing NaN values with the mean for numerical columns, 
# filtering out outliers, encoding categorical values(one-hot encoding) and data normalization(min-max scaling)
```
